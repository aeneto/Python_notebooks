```python
df_pessoas = df_dados[['cod_pessoa',  'genero', 'data_nascimento']].drop_duplicates()
df_clima = df_dados[['clima']].drop_duplicates()
df_hobbie = df_dados[['hobbies']].drop_duplicates()
df_animal_estimacao = df_dados[['animal_estimacao']].drop_duplicates()
df_bebida = df_dados[['bebida_favorita']].drop_duplicates()
```

```python
#INSERINDO DADOS NA TABELA pessoa com o dataframe df_pessoas

for pessoa in df_pessoas.itertuples():
    cod_pessoa = pessoa.cod_pessoa
    genero = pessoa.genero
    data_nascimento = str(pessoa.data_nascimento)
    try:
        query = f"""insert into pessoa (cod_pessoa, genero, data_nascimento ) 
                     values ('{cod_pessoa}',  '{genero}', '{data_nascimento}')"""      
        conn.execute(query)      
        print(f"Registro inserido com sucesso: {genero}")
    except Exception as e:        
        print(f"Não foi possível inserir o regristro {genero}. O erro foi encontrado foi: {e}")   
```

```python
#INSERINDO DADOS NA TABELA pessoa com o dataframe df_clima

for clima in df_clima.itertuples():

    clima = clima.clima
    
    try:
        query = f"""insert into clima ( clima ) 
                     values ( '{clima}')"""      
        conn.execute(query)      
        print(f"Registro inserido com sucesso: {clima}")
    except Exception as e:        
        print(f"Não foi possível inserir o regristro {clima}. O erro foi encontrado foi: {e}")   
```

```python
#INSERINDO DADOS NA TABELA pessoa com o dataframe df_animal_estimacao

for animal in df_animal_estimacao.itertuples():

    animal_estimacao = animal.animal_estimacao
    
    try:
        query = f"""insert into animal_estimacao ( animal ) 
                     values ( '{animal_estimacao}')"""      
        conn.execute(query)      
        print(f"Registro inserido com sucesso: {animal_estimacao}")
    except Exception as e:        
        print(f"Não foi possível inserir o regristro {animal_estimacao}. O erro foi encontrado foi: {e}")   
```

```python
#INSERINDO DADOS NA TABELA pessoa com o dataframe df_hobbie

for hobbie in df_hobbie.itertuples():

    hobbies = hobbie.hobbies
    
    try:
        query = f"""insert into hobbie ( hobbie) 
                     values ( '{hobbies}')"""      
        conn.execute(query)      
        print(f"Registro inserido com sucesso: {hobbies}")
    except Exception as e:        
        print(f"Não foi possível inserir o regristro {hobbies}. O erro foi encontrado foi: {e}")   
```

```python
#INSERINDO DADOS NA TABELA pessoa com o dataframe df_hobbie

for bebida in df_bebida.itertuples():

    bebida_favorita = bebida.bebida_favorita
    
    try:
        query = f"""insert into bebida ( bebida) 
                     values ( '{bebida_favorita}')"""      
        conn.execute(query)      
        print(f"Registro inserido com sucesso: {bebida_favorita}")
    except Exception as e:        
        print(f"Não foi possível inserir o regristro {bebida_favorita}. O erro foi encontrado foi: {e}") 
```
