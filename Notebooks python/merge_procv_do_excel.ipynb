```python
import pandas as pd
import os
```

```python
arquivo_2018 = pd.read_excel('POP2018_20210331.xls')
arquivo_2019 = pd.read_excel('POP2019_20210331.xls')
arquivo_2020 = pd.read_excel('POP2020_20210331.xls')
ibge_censo_amostra_caracteristica_populacao_2 = pd.read_excel('ibge_censo_amostra_caracteristica_populacao_2.xlsx')
```

```python
#unir dois arquivos
df  = pd.merge(arquivo_2018, arquivo_2019, on='codigo', how='left')
```

```python
#unir mais um arquivo
final = pd.merge(df, arquivo_2020, on='codigo', how='left')
```

```python
#unir mais um arquivo
final_2 = pd.merge(final, ibge_censo_amostra_caracteristica_populacao_2, on='codigo', how='left')
```

```python
final_2.columns = [col.replace(",", "") for col in final_2.columns]
final_2
```
