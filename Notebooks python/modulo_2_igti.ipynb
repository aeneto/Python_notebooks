```python
#verificando o número de países diostintos na base
dataframe['country'].nunique()
#o nunique seria o distinct do sql
```

```python
#Quais são os países com maior número de pessoas imunizadas?

# 1 identificando a data mais recente da coleta

ultimadata = dataframe['date'].max()
ultimadata
```

```python
#filtrando a ultima data para trazer todos os registros. o dataframe recebe o valor da ultima data

ultimadata = dataframe[dataframe.date == ultimadata]
```

```python
# o comando sape mostra a quantidade de linhas e colunas dentro de um dataset
ultimadata.shape
```

```python
#nem todos os paises reportaram a vacinação, por isso precisa fazer outro filtro, na qual agrupa os paises e seleciona a ultima data de cada país.
dataframe.groupby('country')['date'].max().sort_values(ascending=False)
```

```python
#Agora vimos que existem países de fato que não possuem a atualização da ultima data da coleta. Dessa forma, temos que procurar uma outra maneira para obter esses dados. A variável total_vaccinations contém o total de vacinações naquele país. Ouse ja, podemos utilizar ele como parâmetro para buscar a última data de coleta. Desta forma, vamos agrupar por país e pegar o valor máximo obtido nesa variável.


dataframe.groupby('country')['total_vaccinations'].max().sort_values(ascending=False)
```

```python
imunizados_por_pais = dataframe.groupby('country')['total_vaccinations'].max().sort_values(ascending=False)
imunizados_por_pais
```

```python
type(imunizados_por_pais)
```

```python
#efetuar uma operação, o seu resultado se transforma em uma série pandas
# uma série pandas se caracteriza por um array unidimensional capaz de armazenar qualquer tipo de dados com rótulos ou índice de eixo. Um exemplo de um objeto Series é uma coluna de um DataFrame


imunizados=imunizados_por_pais.to_frame().reset_index()
imunizados
```

```python
#Respondendo a pergunta. TOP 10

imunizados.sort_values('total_vaccinations',ascending=False).head(10)
```

```python
#Qual vacina é a mais utilizada no mundo ?


vacinas_por_pais = dataframe[['country','date','vaccines']]
vacinas_por_pais
```

```python
#aGORA VAMOS AGRUPAR AS VACINAS E CONTAR OS PAIS APENAS UMA VEZ. SE NO DATASET TIVER UM PAÍS UTILIZANDO DUAS VACINAS, ELE VAI SER CONTATO DUAS VEZES

vacinas_por_pais.groupby('vaccines')['country'].nunique().sort_values(ascending=False)
```

```python
vacina_pais = dataframe.groupby('country')['vaccines'].unique()
vacina_pais = vacina_pais.to_frame().reset_index()
```

```python
vacina_pais
```

```python
#3 - Qual o país possui o maior número de vacinados por dia ? 
dataframe.groupby('country')['daily_vaccinations'].max().sort_values(ascending=False)
```

```python
media_pais = dataframe.groupby('country')['daily_vaccinations'].mean().sort_values(ascending=False)
media_pais = media_pais.to_frame().reset_index()
media_pais

#mean() é a média
```

```python
#Renomeando novas colunas das datas

media_pais = media_pais.rename(columns={'country': 'país', 'vaccines' :'vacina_utilizada'})
media_pais[:5]
```

```python
vacina_pais = vacina_pais.rename(columns={'country': 'país', 'vaccines' :'vacina_utilizada'})
vacina_pais[:5]
```

```python
imunizados = imunizados.rename(columns={'country': 'país', 'vaccines' :'vacina_utilizada'})
imunizados[:5]
```

```python
#Criando um novo dataset
#utiliza-se a função merge. Essa função vai comprar os campos das tabelas que são chaves iguais, como o nome do pais e realizar uma junção entre as tabelas

dados_tratados = pd.merge(imunizados, media_pais, how='inner')

dados_tratados = pd.merge(dados_tratados, vacina_pais)
dados_tratados
```

```python
#exportando dados tratados

dados_tratados.to_csv('estatistica_pais.csv', index=True, sep = ';')
```

```python
dados_tratados.to_json('estatistica.json', orient = 'records')
```
