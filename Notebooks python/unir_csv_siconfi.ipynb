```python
import pandas as pd
import os
import csv
```

```python
#atribuir variável com o diretório onde está localizado o arquivo
DIRETORIO = "C:\\Users\\André Neto\\Documents\\Zetta\\Projetos\\Vale\\SICONFI\\Geral\\Receitas Orcamentarias Previsao Inicial"
```

```python
#Agora vamos criar uma lista com todos os arquivos que estão armazenados no diretório.

#o metódo listdir retorna uma lista contendo os nomes das entradas no diretório fornecido por path.

lista_df = [] # cria uma lista
lista_path = os.listdir(DIRETORIO) 
```

```python
lista_path[:]
```

```python
lista_df = []
for arquivo in lista_path:
    local_arquivo = os.path.join(DIRETORIO, arquivo)
    df = pd.read_csv(local_arquivo, sep = ';', encoding='latin-1')
    lista_df.append(df)
df_arquivo = pd.concat(lista_df) # Atribui os dados coletados em um dataframe
```

```python
df_arquivo[:10]
```

```python
#exportar apenas para um arquivo único
df_arquivo.to_csv('Receitas Orcamentarias Previsao Inicial.csv', sep = ';', encoding='latin-1', index=False)
```
