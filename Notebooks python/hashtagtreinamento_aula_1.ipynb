```python
import pyautogui #faz a automacao do mouse e do teclado
import time #controla o tempo do nosso programa
import pyperclip #ela permite a gente copiar e colar com o python


pyautogui.PAUSE = 1
pyautogui.alert("O programa está rodando")

#Passo 1: Entrar no sistema (Link do Google Drive)
pyautogui.hotkey('ctrl', 't')
link = "https://drive.google.com/drive/folders/1wRTFw0sUVBjRr4hW5U9LF7DjLixRyxym?usp=sharing"

#da ctrl+c 
pyperclip.copy(link)

#ctrl+v
pyautogui.hotkey('ctrl', 'v')
pyautogui.press("enter")

#sperar um pouco
time.sleep(5)

#Passo 2: Entrar na pasta da Aula 1
#pyautogui.click(935, 669, clicks=2)

#Passo 3: Fazer o download da Base de Vendas
pyautogui.click(909,366)
pyautogui.click(1710,151)
pyautogui.click(1528,520)
time.sleep(10)


#Se quiser abrir o navegador ou qualquer programa
# pyautogui.press("win")
#pyautogui.write("chrome")
#pyautogui.press("enter")
```
