```python
dados_ausentes = df_jogos[pd.isnull(df_jogos).any(axis=1)]
```

```python
dados_ausentes[:15]
```

```python
#Este comando alterar tudo de nulos para Dado_ausente

dados_ausentes['num_vitorias'].fillna('Dado_ausente')
```

```python
dados_ausentes['num_vitorias'].fillna(dados_ausentes['total_partidas'] - dados_ausentes['num_derrotas'])
```

```python
#Vamos atribuir ao data_frame
df_jogos['num_vitorias'].fillna(df_jogos['total_partidas'] - df_jogos['num_derrotas'])
```

```python
#Vamos atribuir ao data_frame
df_jogos['num_derrotas'].fillna(df_jogos['total_partidas'] - df_jogos['num_vitorias'])
```

```python
#depois de todas as alterações, devemos atribuir a alteração na coluna

df_jogos['num_vitorias'] = df_jogos['num_vitorias'].fillna(df_jogos['total_partidas'] - df_jogos['num_derrotas'])
df_jogos['num_derrotas'] = df_jogos['num_derrotas'].fillna(df_jogos['total_partidas'] - df_jogos['num_vitorias'])
df_jogos['total_partidas'] = df_jogos['total_partidas'].fillna(df_jogos['num_vitorias'] + df_jogos['num_derrotas'])
```
