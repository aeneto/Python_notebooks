```python
df_pessoas = df_vendas[['cod_pessoa', 'nome', 'genero','data_nascimento']].drop_duplicates() #aqui eu pego somente dados de pessoas
df_modelos = df_vendas[['modelo','fabricante']].drop_duplicates() #aqui dados de modelo e fabricante
fabricantes = list(df_vendas.fabricante.unique()) #neste caso criou-se uma lista, uma vez que os dados contidos são apenas uma coluna
```

```python
#INSERINDO DADOS NA TABELA pessoa com o dataframe df_pessoas

for pessoa in df_pessoas.itertuples():
    cod_pessoa = pessoa.cod_pessoa
    nome = pessoa.nome
    genero = pessoa.genero
    data_nascimento = str(pessoa.data_nascimento)[:10] #para retirar o 
    try:
        query = f"""insert into pessoa (cod_pessoa, nome, genero, data_nascimento)
                values ('{cod_pessoa}', '{nome}', '{genero}', '{data_nascimento}')"""
        conn.execute(query)
        print(f"Registro inserido com sucesso: {nome}")
    except Exception as e:
              print(f"Não foi possível inserir o registro {nome}. O erro encontrado foi: {e}")


```

```python
#INSERINDO DADOS NA TABELA pessoa com o dataframe df_pessoas

for fabricante in fabricantes: #fabricante é uma lista, então vamos percorrer a lista
    try:
        query = f"""insert into fabricante (nome_fabricante)
                    values ('{fabricante}')"""
        conn.execute(query)
        print(f"Registro inserido com sucesso: {fabricante}")
    except Exception as e:
        print(f"Não foi possível inserir registro {fabricante}^. O erro encontrado foi: {e}")
```

```python
# As demais tabelas possuem informações de foreign key, então vamos salvar o select de fabricante em um dataframe

df_fabricante = pd.read_sql('fabricante',con=conn) #Le o dado na tabela do banco de dados
df_fabricante
```

```python
#Setor o index 
dict_fabricante = df_fabricante.set_index(df_fabricante.nome_fabricante)['cod_fabricante'].to_dict()
dict_fabricante

```

```python
#Buscar nome dos fabricantes e associar a seu respectivo codigo
df_modelos.fabricante =  df_modelos.fabricante.map(dict_fabricante)
df_modelos
```

```python
df_modelos = df_modelos.rename(columns={'fabricante': 'cod_fabricante'})
df_modelos
```

```python
##Agora vamos fazer o insert no banco da tabela modelo

for modelo in df_modelos.itertuples():
    nome_modelo = modelo.modelo
    cod_fabricante = modelo.cod_fabricante
    try:        
        query = f"""insert into modelo_veiculo (descricao_modelo, cod_fabricante) 
                     values ('{nome_modelo}',{cod_fabricante})"""
        conn.execute(query)
        print(f"Registro inserido com sucesso: {nome_modelo}")
    except Exception as e:
        print(f'Não foi possível inserir o registro: {nome_modelo} Erro: {e}')
    
```
