```python
import pandas as pd
```

```python
data = pd.read_csv('tabela.csv', sep =';')
#data = pd.read_excel('tabela.xlsx')
```

```python
data
```

```python
"""
indexar colunas e passar as demais para linhas
Exemplo: indexar codigo_municipio e nome_municipio
passar uma coluna para anos, e uma pra valor
"""

data = pd.melt(data.reset_index(), id_vars=['codigo_municipio', 'nome_municipio'], 
var_name='ano', value_name='valor')
```

```python
data
```

```python
#retirar os index e da uma limpada no arquivo
data = data[data.ano.str.contains("index") == False]
data.head(50)
```
