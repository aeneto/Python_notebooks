```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#carregando a base de dados
base = pd.read_csv('sinistralidade.csv')
base.head()
```

```python
#grafico de dispersão considerando o volume e a dispersao
plt.scatter(base.Girth, base.Volume, color = 'blue', facecolor = 'none', marker = '*')
plt.title('Arvores')
plt.xlabel('Volume')
plt.ylabel('Circunferencia')
```

```python
#grafico de dispersão considerando o volume e a dispersao
plt.plot(base.Girth, base.Volume)
plt.title('Arvores')
plt.xlabel('Volume')
plt.ylabel('Circunferencia')
```

```python
#grafico de dispersão com afastamento de dados
sns.regplot(base.Girth, base.Volume, data = base, x_jitter = 0.3, fit_reg = False)
#o x_jitter é como se fosse um zoom
```
